```python
def get_temp_data():
    time = [0, 9, 24+15, 24+23, 48+9, 48+15, 48+23, 72+9, 72+17, 72+23, 96+9, 96+23, 120+8, 120+10, 120+17, 120+23, 144+8, 144+18, 168+9, 168+23, 192+8, 192+18, 192+23, 216+9, 216+14]
    temp = [35.95, 36.50, 36.60, 36.17, 36.50, 37.09, 37.24, 36.53, 37.25, 36.79, 36.59, 
        36.62, 36.45, 36.8, 36.99, 36.19, 36.4, 37.01, 36.5, 35.7, 36.4, 36.98, 36.59, 36.65, 36.87]

    X_train = np.array(time).reshape(-1, 1)
    Y_train = tf.constant(temp, dtype=tf.float32, shape=(len(temp), 1))
    X_test = np.arange(240, 700, 10.).reshape(-1, 1)

    scaler = MinMaxScaler(feature_range=(0, 1))
    X_train = tf.constant(scaler.fit_transform(X_train.reshape(-1, 1)), dtype=tf.float32)
    X_test = tf.constant(scaler.transform(X_test.reshape(-1, 1)), dtype=tf.float32)

    return X_train, Y_train, X_test
```

```python
# define the activations and their corresponding labels
activations = ["tanh","relu"]
labels = ["tanh","ReLU"]

# define the architecture of the model
ARCHITECTURE = [1, 64, 64, 1]
EPOCHS = 3000

# create a simple MLP model
def create_model(activation):
    model = tf.keras.Sequential([
        tf.keras.layers.InputLayer(input_shape=(1,)),
        tf.keras.layers.Dense(units=ARCHITECTURE[1], activation=activation),
        tf.keras.layers.Dense(units=ARCHITECTURE[2], activation=activation),
        tf.keras.layers.Dense(units=ARCHITECTURE[3])
    ])
    return model

# #define the training function
def train_extrapolation(model, X_train, y_train, epochs, verbose):
    model.compile(optimizer=tf.keras.optimizers.Adam(),
                  loss=tf.keras.losses.MeanSquaredError())
    history = model.fit(X_train, y_train, epochs=epochs, verbose=verbose)
    train_losses = history.history["loss"]
    return model, train_losses

# # generate the training and testing data
# X_train_temp = np.arange(0, 720, 10).reshape(-1, 1)
# y_train_temp = 37 * np.sin(np.pi * X_train_temp / 288) + np.random.normal(0, 0.1, size=(len(X_train_temp), 1))
# X_test_temp = np.arange(720, 1440, 10).reshape(-1, 1)

# plot the training data
plt.figure(figsize=(10, 5))
plt.ylim(36, 38)
plt.plot(Y_train_temp, 'x', label="Training set")

# train the model with different activations and plot the predictions
for activation, label in zip(activations, labels):
    model = create_model(activation)
    model, train_losses = train_extrapolation(model, X_train_temp, Y_train_temp, epochs=EPOCHS, verbose=False)
    preds_train = model.predict(X_train_temp)
    preds_test = model.predict(X_test_temp)
    
    plt.plot(np.vstack((preds_train, preds_test)), label="prediction, " + label)

# set the x-axis ticks and labels
plt.xticks(range(0, 71, 9), range(0, 700, 96))
plt.xlabel("Hour")
plt.ylabel("Body Temperature")
plt.legend()
plt.show()
```
